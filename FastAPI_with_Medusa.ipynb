```python
# %%writefile app.py
# from medusa.decoding import MedusaDecoder
import asyncio
from queue import Queue
from fastapi import FastAPI
from pydantic import BaseModel
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

app = FastAPI()

# Load Tokenizer & Model
model_name = "lmsys/vicuna-7b-v1.3"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16).cuda()

# Load Medusa Head for Speculative Decoding
# medusa = MedusaDecoder(model)
from medusa.model.medusa_model import MedusaModel
medusa = MedusaModel(model)
# Batch Processing Queue
request_queue = Queue()

class RequestData(BaseModel):
    prompt: str
    max_tokens: int = 50

# Dynamic Batching Worker
async def batch_worker():
    while True:
        if not request_queue.empty():
            batch = []
            while not request_queue.empty():
                batch.append(request_queue.get())

            prompts = [req["prompt"] for req in batch]
            max_tokens = max(req["max_tokens"] for req in batch)

            # Encode inputs
            inputs = tokenizer(prompts, return_tensors="pt", padding=True).to("cuda")

            # Speculative Decoding with Medusa Head
            with torch.no_grad():
                output = medusa.generate(**inputs, max_new_tokens=max_tokens)

            responses = [tokenizer.decode(out, skip_special_tokens=True) for out in output]

            # Send responses
            for i, req in enumerate(batch):
                req["future"].set_result({"response": responses[i]})

        await asyncio.sleep(0.1)  # Adjust batch frequency

@app.post("/generate")
async def generate_text(request: RequestData):
    loop = asyncio.get_event_loop()
    future = loop.create_future()
    request_queue.put({"prompt": request.prompt, "max_tokens": request.max_tokens, "future": future})
    return await future



# Expose API using ngrok

```
